### <font color=red><b>복합 토큰화 Tokenizer() 1D

```python
from keras.preprocessing.text import Tokenizer    # 복잡한 문장을 토큰화 할 경우 Tokenizer 객체를 가져와 사용
text_list = ['홍길동은 딥러닝을 공부합니다.',
        '홍길동은 딥러닝이 어렵습니다.', 
        '아무개도 딥러닝을 시작합니다.']  # 1차여야 함

token = Tokenizer()
token.fit_on_texts(text_list)   # like fit_transform
print(' keras : \n')
print('문장 카운트 : ',token.document_count) # 몇문장이 있는가?

print("==="*20)
print("문장 토큰화(texts_to_sequences) : ")
text_seq_list = token.texts_to_sequences(text_list)  # 한문장 씩 토큰화한 결과 숫자로 표현 : 문장 토큰화(정수인코딩), 빈도수
print(text_seq_list)

print("==="*20)
for seq, sent in zip (text_seq_list,text_list) :  # zip: 같은 인덱스의 위치에 있는 것끼리 튜플로 묶음  
                                                  # zip([원본문장1,원본문장2,원본문장3],[토큰화1,토큰화2,토큰화3])
                                                  # (원본문장1, 토큰화1)
                                                  # (원본문장2, 토큰화2)
                                                  # (원본문장3, 토큰화3)
    print(sent, seq)

print("==="*20)      
print('단어 카운트 : \n', token.word_counts)   # 등장한 각 단어의 수

print("==="*20)
print('각 단어에 매겨진 인덱스 값 :  \n',token.word_index)   # 해당하는 단어의 인덱스, 자르는 순서로 인덱스 부여

print("==="*20)
print('각 단어가 몇 개의 문장에 포함되어 있는가 :  \n',token.word_docs)    # 문장전체에서의 단어 빈도 수 , sorting
```

```python
corpus = [['This is a black cat'],
          [ 'This is a white dog at home'] ] 
# 
# corpus = np.array(corpus)
# print(corpus, corpus.reshape(-1))


# -------------------- 1D --------------------------
corpus = np.array(corpus).reshape(-1)  
print(corpus)


tokenizer = Tokenizer() # 1차만 받음
tokenizer.fit_on_texts(corpus)
print('정수 인코딩 : ',tokenizer.word_index) 

encoding_txt = tokenizer.texts_to_sequences(corpus)   #[[ 1, 2, ..] , ['1, 2, ..]]
print('인코딩 문장 ; ', encoding_txt)

# ------------------------------------------------------
# #voca_size :문장들이 가지고 있는 단어갯수 중 최대 단어수
max_length = max(len(x.split()) for x in corpus )
print('최대길이(Voca Size):', max_length)


padding_res = pad_sequences(sequences=encoding_txt,   #[[ 1, 2, ..] , ['1, 2, ..]] : 숫자로 되어 있어야 함(인코딩 된 형태의 list)
                   maxlen=10,      #voca_size :문장들이 가지고 있는 단어갯수 중 최대 단어수
                   padding="pre"  #앞에서부터 0을 채워서 제일 긴 길이에 맞추기 pre/post
                  )
print(padding_res)
```

```python
def my_reg(text):
    # text = re.sub(r'/^\s+|\s+$/g', '', text.strip())#앞뒤 공백 제거
    text = text.strip()
    text = text.replace(r' /','/')  #앞뒤 공백 제거
    text = text.replace(r'/ ','/')  #앞뒤 공백 제거
    
    menu_list = []
    for menu in text.split():
        menu = re.sub(r'\([^)]*\)', '', menu.strip())   #(s) 제거
        menu = re.sub(r'[^)]*\)', ''  , menu.strip())     #s) 제거
        menu = re.sub(r'\([^)]*', ''  , menu.strip())     #(s 제거
        menu = re.sub(' +', ' '       , menu.strip())          #공백,탭 제거
        if len(menu) > 0:
            menu_list.append(menu)
    # return menu_list
    return ' '.join(menu_list)


df['중식메뉴2'] = df['중식메뉴'].apply(lambda x: my_reg(x))
df[['중식메뉴','중식메뉴2']].head()
```

- 필요없는 단어 뽑기

```python
from nltk.tokenize import RegexpTokenizer

stop_word_list=[]

for text in df['중식메뉴'] :
    tokenizer1 = RegexpTokenizer("[(][\w]+[\W][\w]+[\W]*[\w]+[)]")
    stop_word=tokenizer1.tokenize(text)
    stop_word_list.append(stop_word)
print(stop_word_list[:5])
```

```python
stop_word_list2=[]
for i in range(len(stop_word_list)) :
    # print(text)
    for j in range(len(stop_word_list[i])) :
        tokenizer1 = RegexpTokenizer("[\w]+")
        stop_word=tokenizer1.tokenize(stop_word_list[i][j])
        stop_word_list2.extend(stop_word)
print(stop_word_list2[:5])
```

```python
from collections import Counter
vocab = Counter(stop_word_list2)
print(vocab)
```
