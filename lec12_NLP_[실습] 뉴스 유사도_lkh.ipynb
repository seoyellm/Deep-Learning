```python
import pandas as pd
```

```python
df = pd.read_csv('./datasets/craw_ytn_news.csv')
```

```python
df.isna().sum()
```

```python
isna_idx = df['content'].isna()[df['content'].isna()].index.tolist()
```

```python
for idx in isna_idx:
    df.drop(index= idx, inplace=True)
print(df.isna().sum())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf= TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['content'])
print(tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(cosine_sim)
```

```python
cos_df = pd.DataFrame(cosine_sim, columns=df["title"], index=df["title"])
cos_df.head()
```

```python
search_text = "누리호"
search_list = cos_df[cos_df.index.str.contains(search_text)].index
print(f"검색 결과 뉴스 : {len(search_list)}건\n")
for title_str in search_list[:5]:
    print(title_str)
```
